```python
import pandas as pd
```

```python
df = pd.read_csv("non-ribosomal lipopeptides - list - Sheet2.csv")
```

```python
df_nonull = df[df["AMINO ACID STARTER CONDENSATION DOMAIN"].notnull()]
df_starter = df_nonull[df_nonull["AMINO ACID STARTER CONDENSATION DOMAIN"].str.match('[A-Z]')]
```

```python
df_starter
```

```python
starter_condensation_domains = df_starter["AMINO ACID STARTER CONDENSATION DOMAIN"].values
BGCs = df_starter["Fasta File"].values
```

```python
BGCs
```

```python
fasta_entries = list(zip(BGCs, starter_condensation_domains))
```

```python
with open("collective_starters.fasta", "w") as f:
    for BGC, starter in fasta_entries:
        f.write(f">{BGC}\n{starter}\n")
```
