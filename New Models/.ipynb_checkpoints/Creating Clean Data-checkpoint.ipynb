```python
import pandas as pd
import re
import numpy as np
```

```python
raw_data = pd.read_csv("lipopeptide_data.csv")
```

```python
raw_data
```

```python
dropped_data = raw_data.drop(raw_data.columns[[0,1,2,4,10,14,15]], axis = 1)
```

```python
dropped_data_na = dropped_data.dropna(subset = ["AMINO ACID STARTER CONDENSATION DOMAIN"])
```

```python
dropped_data_na
```

```python
processed_df = dropped_data_na.copy()
processed_df.dropna(subset=["CHAIN LENGTH"], inplace = True)
for i in processed_df.index:
    if re.search(r"^\d+", processed_df.at[i,"CHAIN LENGTH"]) is None:
        processed_df.at[i, "CHAIN LENGTH"] = 0 
    else:
        processed_df.at[i, "CHAIN LENGTH"] = 1
processed_df["RzmA H139, H140, D144 (Conserved motif)"] = processed_df["RzmA H139, H140, D144 (Conserved motif)"].fillna("0")
```

```python
clean_data = pd.DataFrame()
clean_data[["# Amide bonds", "AMINO ACID STARTER CONDENSATION DOMAIN"]] = processed_df[["# Amide bonds", "AMINO ACID STARTER CONDENSATION DOMAIN"]]
clean_data["RzmA H139, H140, D144"] = ""
```
